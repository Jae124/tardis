```python
import os
import logging
from carsus.io import AtomDataCompare
```

```python
from carsus.io.output import TARDISAtomData
```

## NIST data

```python
from carsus.io.nist import NISTWeightsComp, NISTIonizationEnergies
```

```python
atomic_weights = NISTWeightsComp()
```

```python
ionization_energies = NISTIonizationEnergies('H-Zn')
```

## GFall data

```python
!wget -qO /tmp/gfall.dat https://media.githubusercontent.com/media/tardis-sn/carsus-db/master/gfall/gfall_latest.dat
```

```python
from carsus.io.kurucz import GFALLReader

gfall_reader = GFALLReader('H-Zn', '/tmp/gfall.dat')
```

## ZETA data

```python
from carsus.io.zeta import KnoxLongZeta

zeta_data = KnoxLongZeta()
```

## Chianti v. 9.0.1

```python
from carsus.io.chianti_ import ChiantiReader
```

```python
chianti_reader = ChiantiReader('H-Si', collisions=True, priority=20)
```

```python
v9_0_data=TARDISAtomData(atomic_weights, ionization_energies, gfall_reader,zeta_data,chianti_reader,cmfgen_reader=None)
```

```python
v9_0_data.to_hdf('version9.0.h5') ## creating the data file
```

## Comparing files

```python
ATOM1_PATH = "version9.0.h5"
ATOM2_PATH = "version10.0.h5"
```

```python
atc = AtomDataCompare(ATOM1_PATH, ATOM2_PATH)
```

```python
atc.compare()
```

```python
atc.key_diff("levels")
```

```python
atc.ion_diff("levels",(8,2))
```

```python
atc.key_diff("lines")
```

```python
atc.ion_diff("lines", (8,0))
```

```python
atc.ion_diff("lines","C",return_summary=True)
```
